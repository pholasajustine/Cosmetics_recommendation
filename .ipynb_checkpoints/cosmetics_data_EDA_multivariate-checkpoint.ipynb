#### Let's create a colum to have the proportion of harmful ingredients

```python
data["Harmful_part"] = data.Harmful_Substances / data.Ingredients.str.len() * 100
```

```python
data.head()
```

```python
data.describe()
```

On average, there are 0.5% harmful ingredients in all cosmetic products.
In view of the lower median, it seems that some products have more harmful ingredients.
The maximum toxicity proportion of the product is 2.88%
There are products that contain 0 harmful ingredients (in the data cleaning we saw, there were 257).

## Analysis of the "safe" products that don't contain any harmuful susbtances:

```python
data_safe = data.query('Harmful_part == 0')
```

```python
# 15% products are "safe"
len(data_safe) / len(data) * 100
```

```python
data_safe['Price'].hist(bins=50)
```

```python
data_others = data.query('Harmful_part > 0')
```

```python
data_others['Price'].hist(bins=50)
```

The price does not seem to be influenced by the proportion of healthy ingredients because the 2 distributions are alike.

## Toxicity rate by brand 

```python
plt.figure(figsize=(50, 20))
ax = sns.boxplot(x="Brand", y="Harmful_part", data=data)
ax.set_xticklabels = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

In terms of the number of ingredients per product, it seems that certain brands are to be avoided / preferred.

Brands to Avoid: Shiseido; Lancôme, Laura Mercier, Becca, Naturally Serious

> However, we must see the harmful ingredients not in terms of quantity, but of high toxicity. What ingredients to avoid? 

## Price per brand

```python
plt.figure(figsize=(50, 20))
ax = sns.boxplot(x="Brand", y="Price", data=data)
ax.set_xticklabels = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

In terms of product prices by brand, the prices of La Mer & Guerlain products are very variable and can be the most expensive.
There are some very expensive price outliers at some brands.

## Toxicity rate by product type 

```python
data.Label.value_counts()
```

```python
plt.figure(figsize=(50, 20))
ax = sns.boxplot(x="Label", y="Harmful_part", data=data)
ax.set_xticklabels = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

The number of harmful ingredients per product obviously does not depend on the type of product. However, it seems that the cleaning products are more concentrated in number of harmful ingredients in view of the outliers. 

## Price per type of product

```python
plt.figure(figsize=(50, 20))
ax = sns.boxplot(x="Label", y="Price", data=data)
ax.set_xticklabels = ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
```

Prices seem to be influenced by the type of cosmetic product. We note that treatments and creams are generally more expensive than other types.
